#### **What factors have the most impact on the overall score?**

While it sounds quite obvious that the academic reputation has a clear impact on the overall score, the point here is that this is the variable with the strongest linear correlation with the overall score.

##### **3) Sustainability and Employment outcomes (SUS and EO)** = 0.67 and 0.65 respectively. These two variables are the third and fourth variables in the list of strongest linear correlation, meaning that environmental and social impact initiatives and how well graduates perform in the job market have a very strong impact in the overall score.

#### **Other points of interest**

##### **ISR vs. ISD:** International student ratio (ISR, percentage of international students) and international student diversity (ISD, diversity among international students) are almost perfectly correlated. Also, both have a moderate relationship with the overall score. It might be worth digging deeper to discover whether or not they're measuring very similar things.

##### **CPF:** I would expect citations per faculty (research output and quality) to have a stronger relationship than 0.5, given the importance that research has on the prestige of universities. Perhaps its impact varies per region, size or focus?

### **6) Categorical plot**

##### Create a categorical variable that splits the 'Overall SCORE' column into categories

```python
sns.histplot(df_2026_ranking['Overall SCORE'], bins = 20, kde = True)
```

##### Taking as reference the grading system [most commonly used in the US](https://en.wikipedia.org/wiki/Academic_grading_in_the_United_States), I propose the following categories:

##### Tier 1: > 89
##### Tier 2: 80 - 89
##### Tier 3: 70 - 79
##### Tier 4: 60 - 69
##### Tier 5: < 60

##### **Why the word 'Tier'?**

##### I looked for a neutral term - as the purpose of this categorization is for visualization purposes and not to use language with negative connotations.

```python
df_2026_ranking.loc[df_2026_ranking['Overall SCORE'] > 89, 'Score Category'] = 'Tier 1'
```

```python
df_2026_ranking.loc[(df_2026_ranking['Overall SCORE'] >= 80) & (df_2026_ranking['Overall SCORE'] <= 89), 'Score Category'] = 'Tier 2'
```

```python
df_2026_ranking.loc[(df_2026_ranking['Overall SCORE'] >= 70) & (df_2026_ranking['Overall SCORE'] <= 79), 'Score Category'] = 'Tier 3'
```

```python
df_2026_ranking.loc[(df_2026_ranking['Overall SCORE'] >= 60) & (df_2026_ranking['Overall SCORE'] <= 69), 'Score Category'] = 'Tier 4'
```

```python
df_2026_ranking.loc[df_2026_ranking['Overall SCORE'] < 60, 'Score Category'] = 'Tier 5'
```

```python
# Let's check how many universities there are within each category

df_2026_ranking['Score Category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="Region", y="Overall SCORE", hue="Score Category", data=df_2026_ranking)
```

##### **Interpretation**

##### This visualization shows that the Americas, Europe, and Asia have similar distributions of universities across all tiers; still, the Americas have the most universities in the tier 1.

##### It is also clear that Africa is the continent with less universities in the ranking - and that they are within tiers 4 and 5.

##### Europe and Asia have a high density of universities in tier 5 - visually higher than the other continents.
